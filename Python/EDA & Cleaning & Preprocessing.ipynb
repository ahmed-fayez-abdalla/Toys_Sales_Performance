### Sales Dataset

#### Adding some features to the dataset like (revenue - profit)

```python
sales_product = pd.merge(sales, products, on='Product_ID')
sales_product
```

#### Add Revenue Column

```python
sales_product['Revenue'] = sales_product.Units * sales_product.Product_Price
sales_product
```

#### Add Profit Column

```python
sales_product['Profit'] = sales_product.Revenue - (sales_product.Product_Cost * sales_product.Units)
sales_product
```

```python
sales_product.head(20)
```

```python
sales = sales_product[['Sale_ID', 'Date', 'Store_ID', 'Product_ID', 'Units', 'Revenue',  'Profit']]
sales
```

### Monthly Units Sales Trend

```python
monthly_sales = sales.groupby(pd.Grouper(key='Date', freq='ME'))['Units'].sum()
monthly_sales
```

```python
# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Monthly Units Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Monthly Revenue Trend

```python
sales_products = pd.merge(sales, products, on='Product_ID', how='left')
sales_products
```

```python
sales_products['Revenue'] = sales_products['Units'] * sales_products['Product_Price']
sales_products
```

```python
monthly_sales_products_revenue = sales_products.groupby(pd.Grouper(key='Date', freq='ME'))['Revenue'].sum()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales_products_revenue.index, y=monthly_sales_products_revenue.values)
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Top 10 Stores By Units Sales

```python
sales_stores = pd.merge(sales, stores, on='Store_ID', how='left')
sales_stores
```

```python
store_sales = sales_stores.groupby('Store_Name')['Units'].sum().sort_values(ascending=False)
store_sales_top_10 = store_sales.iloc[0:10]
store_sales_top_10
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=store_sales_top_10.index, y=store_sales_top_10.values)
plt.title('Top 10 Stores By Units Sales')
plt.xlabel('Store Name')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Revenue By Category

```python
category_revenue = sales_products.groupby('Product_Category')['Revenue'].sum()
category_revenue
```

```python
plt.figure(figsize=(8, 8))
plt.pie(category_revenue, labels=category_revenue.index, autopct='%1.1f%%')
plt.title('Revenue Product Category')
plt.show()
```

### Correlation between Product Price And Stock On Hand

```python
stock_on_hand_per_product = inventory.groupby('Product_ID')[['Stock_On_Hand']].sum().reset_index()
inventory_products = pd.merge(stock_on_hand_per_product, products, on='Product_ID', how='left')
inventory_products
```

```python
sns.regplot(inventory_products, x='Product_Price', y='Stock_On_Hand')
plt.show()
```

```python
# Contrary to expectations, the relation between Stock_on_Hand and Product_Price is an inverse relationship
```

```python
inventory_stores = pd.merge(inventory, stores, on='Store_ID', how='left')
stock_per_location = inventory_stores.groupby('Store_Location')[['Stock_On_Hand']].sum().sort_values(by='Stock_On_Hand').reset_index()
stock_per_location
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=stock_per_location.Stock_On_Hand, y=stock_per_location.Store_Location)
plt.title('Stock Per Location')
plt.xlabel('Stock On Hand')
plt.ylabel('Store Location')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
